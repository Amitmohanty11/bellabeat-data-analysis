Exploratory Data Analysis


```python
df.columns
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(15,5))
sns.lineplot(x='Date',y='WeightKg',data=df)
plt.xlabel('Date')
plt.xticks(rotation=45)
plt.ylabel('Weight(kg)')
plt.title('Weight over time')
plt.show()
```

we can see the increase in weight overtime

```python
sns.set(style='darkgrid')
plt.figure(figsize=(15,6))
sns.lineplot(x='Date',y='BMI',data=df)
plt.xlabel('date')
plt.xticks(rotation=45)
plt.ylabel('BMI')
plt.title('BMI over time')
plt.show()
```

we can see the increase in BMI overtime

```python
sns.set(style='darkgrid')
plt.figure(figsize=(25,10))
sns.barplot(x='WeightKg',y='BMI',data=df)
plt.xlabel('Weight(kg)')
plt.xticks(rotation=45)
plt.ylabel('BMI')
plt.title('BMI vs Weight')
plt.show()
```

this graph shows how body mass vary with wewight

```python
sns.set(style='darkgrid')
plt.figure(figsize=(5,4))
sns.regplot(x='BMI',y='WeightKg',data=df,scatter_kws={'alpha':0.6},line_kws={'color':'Red'})
plt.xlabel('BMI')
plt.ylabel('Weight(kg)')
plt.title('BMI vs Weight')
plt.show()
```

the above graph tells us how much our data depicts accurate BMI to weight ratio all the data point closer to oe on the red line have perfect value while those farther away from the line are outlier

```python
plt.figure(figsize=(7,4))
sns.histplot(df['BMI'], kde=True, bins=20)
plt.title('Distribution of BMI')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.show()
```

The above graph show us about how body mass is spread across different data points,
here frequency or y axis depicts data points


from this graph we can see that most data pint have BMI less than 30 which shows us that the majority of data point in the dataset have a relatively healthy weight some data point are outlier which tells us if the user is overweight or underweight
