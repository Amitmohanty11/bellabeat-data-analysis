```python
df['SleepDay']=df['SleepDay'].astype('datetime64[ns]')
```

```python
df['SleepHours']=df['TotalMinutesAsleep']/60
df['BedHours']=df['TotalTimeInBed']/60
```

```python
sns.set(style='darkgrid')
plt.figure(figsize=(8,6))
sns.lineplot(x='SleepDay',y='TotalMinutesAsleep',data=df)
plt.xlabel('sleep date')
plt.xticks(rotation=45)
plt.ylabel('Total Minutes Asleep')
plt.title('Total minutes asleep over time')
plt.show()
```

```python
sns.set(style='darkgrid')
plt.figure(figsize=(8,6))
sns.lineplot(x='SleepDay',y='TotalTimeInBed',data=df)
plt.xlabel('sleep date')
plt.xticks(rotation=45)
plt.ylabel('TotalTimeInBed')
plt.title('Total time in bed over time')
plt.show()
```

this graph show total bed bed time overtime

```python
sns.set(style='darkgrid')
plt.figure(figsize=(8,6))
sns.scatterplot(x='TotalTimeInBed',y='TotalMinutesAsleep',data=df)
plt.xlabel('TotalTimeInBed')
plt.ylabel('TotalTimeAsleep')
plt.xticks(rotation=45)
plt.title('Total Time In Bed vs Total Time Asleep')
plt.show()
```

```python
sns.set(style='darkgrid')
plt.figure(figsize=(8,6))
sns.regplot(x='TotalTimeInBed',y='TotalMinutesAsleep',data=df,line_kws={'color':'red'})
plt.xlabel('TotalTimeInBed')
plt.ylabel('TotalTimeAsleep')
plt.xticks(rotation=45)
plt.title('Total Time In Bed vs Total Time Asleep')
plt.show()
```

this plot tells us that the data points falling on the red line are those which have comparable or very close time in bed and sleeping time meaning that those falling in the red line have very less difference in time in bed and sleeping time , but the data pointers far from the red line show that the data point which are off and rent sleeoing well (there is high bed time bus low sleeping time)

```python
sns.set(style='darkgrid')
plt.figure(figsize=(15,8))
sns.histplot(data=df,x='SleepHours',bins=20,kde=True,color='purple')
plt.xlabel('SleepHours')
plt.ylabel('Frequency')
plt.title('Distribution of total Hours asleep')
plt.show()
```

```python
sns.set(style='darkgrid')
plt.figure(figsize=(15,8))
sns.histplot(data=df,x='BedHours',bins=20,kde=True,color='red')
plt.xlabel('BedHours')
plt.ylabel('Frequency')
plt.title('Distribution of total Hours in bed')
plt.show()
```

```python
df_small = df[['SleepDay', 'SleepHours', 'BedHours']].head(15).copy()

df_plot = pd.DataFrame({
    'SleepDay': list(df_small['SleepDay']) * 2,
    'Hours': list(df_small['SleepHours']) + list(df_small['BedHours']),
    'Metric': ['Asleep'] * len(df_small) + ['In Bed'] * len(df_small)
})

plt.figure(figsize=(12,6))
sns.barplot(x='SleepDay', y='Hours', hue='Metric', data=df_plot)
plt.title('Comparison of Time Asleep vs Time in Bed (First 15 Records)')
plt.xlabel('Sleep Day')
plt.ylabel('Hours')
plt.xticks(rotation=45)
plt.show()
```

Hence from this analysis we can see that people are going to bed but there actual sleeping time is less than the time they spend in bed ,from the distribution graphs we can see that how thec frequency of the time spent in bed is higher than time spent sleeping
