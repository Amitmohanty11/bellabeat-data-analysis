```python
sns.set(style='whitegrid')
plt.figure(figsize=(10,6))
ax=sns.lineplot(x='Calories',y='TotalDistance',data=df)
plt.title('Calories vs Total Distance')
plt.xlabel('Calories')
plt.ylabel('Total Distance')
plt.show()
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
plt.figure(figsize=(8,6))
sns.set(style='whitegrid')
ax=sns.scatterplot(x='Calories',y='TotalSteps',data=df,hue='distance')
plt.title('Calories vs Total steps')
plt.xlabel('calories')
plt.ylabel('Total steps')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.set(style='whitegrid')
ax=sns.regplot(x='Calories',y='TotalSteps',data=df,scatter_kws={'alpha':0.5},line_kws={'color':'red'})
plt.xlabel('Calories')
plt.ylabel('Totsl Steps')
plt.title('Calories vs Total steps')
plt.show()
```

This plot suggest taht more steps = more calories burned

This graph show how there is a positive trend in the calories and total steps taken relationship,here each dot discribe a day or the time where data was taken and the red line show us how calories and total steps are related this is shoing a positive trend and the line is moving upward the blue dots on the line are the correct values or normal values and the blue dots far from the red line are unusual data points or the days where calorie count are unexpectedly high or low for the number of steps taken


```python
sns.set(style='whitegrid')
distance_types = ['VeryActiveDistance', 'ModeratelyActiveDistance', 'LightActiveDistance']

plt.figure(figsize=(18, 5))

for i, dist in enumerate(distance_types):
    plt.subplot(1, 3, i + 1)
    sns.regplot(x=dist, y='Calories', data=df,scatter_kws={'alpha': 0.5},line_kws={'color': 'red'})
    plt.title('Calories vs ' + dist)
    plt.xlabel(dist)
plt.show()
```

This plot show how diffent distance effect calories

```python
df['WeekDays']=df['ActivityDate'].dt.day_name()
sns.set(style='whitegrid')
plt.figure(figsize=(8,5))
ax=sns.barplot(x='WeekDays',y='Calories',data=df,color='red',order=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])
plt.xlabel('weekdays')
plt.ylabel('Calories')
plt.title('calories vs weekdays')
plt.show()
```

this shows the calories burned on each day of the week .This also shows on which days of the week the users are readily active
